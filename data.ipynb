```python
import pandas as pd

file_path = "D:/en.openfoodfacts.org.products.csv.gz"

df = pd.read_csv(
    file_path,
    compression="gzip",
    sep="\t",
    engine="python",      # QUAN TRỌNG
    on_bad_lines="skip"   # Bỏ dòng lỗi
)

print(df.shape)



# Danh sách cột cần giữ cho đồ án
selected_columns = [
    # Thông tin sản phẩm
    "code",
    "product_name",
    "brands",
    "categories",
    "main_category",
    "countries_en",

    # Thành phần
    "ingredients_text",
    "additives_n",
    "additives_tags",
    "allergens",

    # Dinh dưỡng (100g)
    "energy-kcal_100g",
    "fat_100g",
    "saturated-fat_100g",
    "carbohydrates_100g",
    "sugars_100g",
    "fiber_100g",
    "proteins_100g",
    "salt_100g",
    "sodium_100g",

    # Chỉ số sức khỏe
    "nutriscore_score",
    "nutriscore_grade",
    "nova_group",
    "nutrition-score-fr_100g",

    # Môi trường
    "environmental_score_score",
    "environmental_score_grade",
    "carbon-footprint_100g"
]

# Lọc dataframe
df_selected = df[selected_columns]

# Xem nhanh dữ liệu
print(df_selected.head())
print(df_selected.info())

# # (Tuỳ chọn) Lưu ra file mới gọn nhẹ hơn
# df_selected.to_csv("openfoodfacts_selected.csv", index=False)
```
